# Introducción a NeuralProphet: **Predicción de series temporales con redes neuronales**

La predicción de series temporales es un desafío importante en el campo de la analítica de datos y el aprendizaje automático. Las series temporales, que son datos secuenciales que varían a lo largo del tiempo, se encuentran en numerosos dominios, como las finanzas, la energía, el clima y las ventas. En este contexto, NeuralProphet ha surgido como una herramienta poderosa y accesible para abordar la predicción de series temporales utilizando redes neuronales.

NeuralProphet es una biblioteca de código abierto desarrollada por la comunidad de Facebook, que combina la simplicidad de uso con la capacidad de modelar y pronosticar series temporales de manera efectiva. Esta herramienta se basa en la arquitectura de las redes neuronales y aprovecha las capacidades de aprendizaje profundo para capturar patrones complejos y hacer predicciones precisas.

Una de las ventajas distintivas de NeuralProphet es su facilidad de uso. A diferencia de otras bibliotecas de aprendizaje automático, NeuralProphet está diseñada para ser accesible incluso para aquellos sin experiencia profunda en ciencia de datos. Su interfaz simple y bien documentada permite a los usuarios cargar y explorar datos de series temporales, ajustar modelos y realizar predicciones en solo unas pocas líneas de código.

Además de su facilidad de uso, NeuralProphet ofrece una amplia gama de características y funcionalidades que lo hacen ideal para la predicción de series temporales. Permite modelar tendencias no lineales, estacionalidad, efectos de días festivos y cambios en la tendencia a lo largo del tiempo. También admite la inclusión de regresores externos, lo que permite tener en cuenta variables adicionales que pueden influir en los datos de la serie temporal.

Otra característica destacada de NeuralProphet es su capacidad para manejar automáticamente múltiples hiperparámetros del modelo, como la selección de la arquitectura de la red neuronal, la regularización y la optimización del aprendizaje. Esto reduce la carga de trabajo para los usuarios y simplifica el proceso de ajuste del modelo, permitiendo un enfoque más eficiente en la exploración y análisis de los resultados.

En resumen, NeuralProphet es una herramienta poderosa y accesible para la predicción de series temporales con redes neuronales. Su enfoque simplificado y su capacidad para capturar patrones complejos hacen que sea una opción atractiva para aquellos que deseen realizar pronósticos precisos en diversos dominios. Al ofrecer una combinación de facilidad de uso, funcionalidades flexibles y automatización de hiperparámetros, NeuralProphet se posiciona como una herramienta valiosa para los profesionales de la ciencia de datos y aquellos interesados en aprovechar el potencial de las redes neuronales en la predicción de series temporales.

# `NeuralProphet:` Un enfoque avanzado para el pronóstico de series de tiempo

El Forecasting preciso de series de tiempo es esencial en diversas industrias y campos, desde las finanzas hasta la logística y la meteorología. En los últimos años, el campo del aprendizaje automático ha visto avances significativos en el pronóstico de series de tiempo, y una de las herramientas más destacadas en este ámbito es NeuralProphet. Con su enfoque basado en redes neuronales, NeuralProphet ha demostrado ser una opción potente y efectiva para aquellos que buscan pronósticos precisos y fiables en datos secuenciales.

NeuralProphet es una librería de Python desarrollada por Facebook Research que se basa en el conocido modelo de Facebook Prophet. A diferencia de Prophet, que utiliza un enfoque aditivo para modelar componentes de series de tiempo como tendencias y estacionalidad, NeuralProphet utiliza redes neuronales para capturar patrones complejos y no lineales en los datos.

Una de las características destacadas de NeuralProphet es su capacidad para manejar automáticamente varios tipos de componentes de series de tiempo, como tendencias, estacionalidad y días festivos. Esto significa que no es necesario especificar manualmente estas componentes, NeuralProphet las identifica y modela de forma automática, lo que ahorra tiempo y esfuerzo en el proceso de pronóstico.

Además, NeuralProphet utiliza la arquitectura de red neuronal conocida como Red Neuronal de Atención (Attention Neural Network), que permite capturar relaciones a largo plazo y depender menos de la longitud de la secuencia de datos. Esto es especialmente beneficioso para las series de tiempo con dependencias temporales complejas y a largo plazo, donde las técnicas tradicionales pueden tener dificultades para capturar patrones adecuadamente.

Otra ventaja de NeuralProphet es su facilidad de uso. La librería proporciona una interfaz intuitiva y amigable para los usuarios, lo que facilita la construcción y el ajuste de los modelos de pronóstico. También ofrece una amplia gama de opciones de configuración, como la selección de hiperparámetros y la personalización de componentes de series de tiempo, lo que permite adaptar el modelo a diferentes escenarios y requisitos específicos.

La implementación de NeuralProphet se ha vuelto más accesible gracias a su disponibilidad como una librería de código abierto. Esto ha permitido que investigadores, científicos de datos y profesionales en general puedan aprovechar las capacidades de NeuralProphet en sus proyectos de pronóstico de series de tiempo. Además, la librería se integra fácilmente con otras herramientas y bibliotecas de aprendizaje automático en el ecosistema de Python, lo que facilita su incorporación en flujos de trabajo existentes.

En resumen, NeuralProphet se ha establecido como una herramienta avanzada y efectiva para el pronóstico de series de tiempo. Su enfoque basado en redes neuronales y su capacidad para modelar automáticamente diferentes componentes de series de tiempo lo convierten en una opción atractiva para aquellos que buscan pronósticos precisos y fiables en datos secuenciales. Con su fácil uso y su disponibilidad como código abierto, NeuralProphet ha abierto nuevas oportunidades en el campo del pronóstico de series de tiempo y ha demostrado su valía en una variedad de aplicaciones prácticas.

## Agregando estacionalidad y autoregresión con Exogena

Los regresores rezagados se utilizan para correlacionar otras variables observadas con nuestra serie de tiempo objetivo. 

A menudo se las denomina variables exógenas o covariables. A diferencia de los futuros regresores, desconocemos el futuro de un regresor rezagado.

En el momento de realizar el pronóstico, solo tenemos acceso a sus valores pasados ​​observados hasta e incluyendo $t-1$

$$\text{Lagged regressor}(t) = L(t) = \sum_{x \in X}L_x(x_{t-1},x_{t-2},...,x_{t-p})$$

```python
m4 = NeuralProphet(
     # Disable trend changepoints
    n_changepoints=10,
    # Disable seasonality components
    yearly_seasonality=True,
    weekly_seasonality=True,
    daily_seasonality=True,
    # Add the autogression
    n_lags=10,)
```
